```python
#base code from: https://www.youtube.com/watch?v=UYgyRArKDEs and the others in series(Keith Galli), although changes made
import numpy as np
import math
import pygame
from BaralTechButtonClass import Button
```

```python
#functions 
row_count = 6
col_count = 7
def create_board():
    board= np.zeros((row_count,col_count))
    return board

def is_valid_location(num):
    if num <= 6 and num>=0:
        return True
    else:
        return False
```

```python
def open_row(boards, colu):
    for i in range(5,-1, -1):
        if boards[i,colu] == 0.:
            global row
            row = i
            return True
    return False
```

```python
def win(board, row, col,turn):
    global game_over
    #horizontal wins
    try:
        board[row+3,col]
        if board[row,col] == board[row+1, col]== board[row+2, col] == board[row+3, col]:
            winning_screen(turn)
            game_over = True
    except IndexError:
        pass
    
    try:
        board[row-3, col]
        if board[row,col] == board[row-1, col]== board[row-2, col] == board[row-3, col]:
            winning_screen(turn)
            game_over = True
    except IndexError:
        pass
   
    #vertical wins
    try:
        board[row, col+3]
        if board[row,col] == board[row, col+1]== board[row, col+2] == board[row, col+3]:
            winning_screen(turn)
            game_over = True
            
    except IndexError:
        pass
    
    try:
        board[row, col-3]
        if board[row,col] == board[row, col-1]== board[row, col-2] == board[row, col-3]:
            winning_screen(turn)
            game_over = True
    except IndexError:
        pass
    
    #diagonal wins
    try: 
        board[row+3, col+3]
        if board[row,col] == board[row+1, col+1]== board[row+2, col+2] == board[row+3, col+3]:
            winning_screen(turn)
            game_over = True
    except IndexError:
        pass
    
    try:
        board[row-3, col-3]
        if board[row,col] == board[row-1, col-1]== board[row-2, col-2] == board[row-3, col-3]:
            winning_screen(turn)
            game_over = True
    except IndexError:
        pass
            
    try:
        board[row+3, col-3]
        if board[row,col] == board[row+1, col-1]== board[row+2, col-2] == board[row+3, col-3]:
            winning_screen(turn)
            game_over = True
    except IndexError:
        pass
    
    try:
        board[row-3, col+3]
        if board[row,col] == board[row-1, col+1]== board[row-2, col+2] == board[row-3, col+3]:
            winning_screen(turn)
            game_over = True
    except IndexError:
        pass
```

```python
#creating board
board = create_board()
game_over= False
turn = 0;
```

```python
def drop_down(pos):
    if pos <= 100:
        return 0
    elif pos <=200:
        return 1
    elif pos <=300:
        return 2
    elif pos <=400:
        return 3
    elif pos <=500:
        return 4
    elif pos <=600:
        return 5
    elif pos <=700:
        return 6
    
```

```python
def winning_screen(turn):
    RED = (236, 0, 0)
    YELLOW = (255, 243, 57)
    pygame.draw.rect(screen, BLACK, (0,0,width,SQUARESIZE))

    #image = pygame.image.load('confetti.png')
    ##display.fill((0,0,0))
    #pygame.display.blit(image,(0,0))

    font = pygame.font.SysFont('comicsansms', 32)

    if turn==0:
        text = font.render('Player 1 - YOU ARE THE WINNER!',True, RED)
        screen.blit(text, (70, 40))
    if turn==1:
        text = font.render('Player 2 - YOU ARE THE WINNER!', True, YELLOW)
        screen.blit(text, (70, 40))

    pygame.display.update()
    
```

```python
#drawing the board
BLUE = (0,127,238)
BLACK = (0,0,0)
RED = (236, 0, 0)
YELLOW = (255, 243, 57)

SQUARESIZE = 100
RADIUS =  SQUARESIZE-55

width = 700
length = 700
size = (width,length)
```

```python
def main_menu(): #Main Menu Screen
    is_running_main_menu = True
    pygame.display.set_caption("Menu")
    
    while(is_running_main_menu):
        
        #screen.fill(BLACK)
        #bg = pygame.image.load("c4.jpg")
        #screen.blit(bg, (0, 0))
        
        menu_mouse_pos = pygame.mouse.get_pos()
        
        two_player_button = Button(image=pygame.image.load("button_2_player.jpg"), x_pos=350,y_pos=300)
        computer_button = Button(image=pygame.image.load("button_computer.jpg"), x_pos=350,y_pos=300)
        two_player_button.update(screen)
        computer_button.update(screen)
        
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit()
                is_running_main_menu = False
            if event.type == pygame.MOUSEBUTTONDOWN:
                if two_player_button.checkForInput(menu_mouse_pos):
                    draw_board(board)
                if computer_button.checkForInput(menu_mouse_pos):
                    draw_board(board)
```

```python
def draw_board(board):
    for c in range(col_count):
        for r in range(row_count):
            pygame.draw.rect(screen,BLUE,(c*SQUARESIZE, r*SQUARESIZE+SQUARESIZE,SQUARESIZE, SQUARESIZE))
            if board[r][c] == 0:
                pygame.draw.circle(screen,BLACK, (c*SQUARESIZE + 50, r*SQUARESIZE+SQUARESIZE+50), RADIUS)
            elif board[r][c] == 1:
                pygame.draw.circle(screen,RED, (c*SQUARESIZE + 50, r*SQUARESIZE+SQUARESIZE+50), RADIUS)
            elif board[r][c] == 2:
                pygame.draw.circle(screen,YELLOW, (c*SQUARESIZE + 50, r*SQUARESIZE+SQUARESIZE+50), RADIUS)
```

```python

def two_player(): #implimenting 2 player game
    
    is_running_two_player = True
    
    #displaying and updating GUI
    screen.fill(BLACK)
    draw_board(board)
    pygame.display.update()
    
    
    while (is_running_two_player): 
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                is_running_two_player = False
            if event.type == pygame.MOUSEMOTION:
                if game_over:
                    turn = 0;
                    board = create_board()
                    draw_board(board)
                    pygame.display.update()
                    game_over = False
                pygame.draw.rect(screen, BLACK, (0,0,width,SQUARESIZE))
                posx = event.pos[0]
                if turn == 0:
                    pygame.draw.circle(screen, RED, (posx, SQUARESIZE/2), RADIUS)
                else:
                    pygame.draw.circle(screen, YELLOW, (posx, SQUARESIZE/2), RADIUS)

                pygame.display.update()

            if event.type == pygame.MOUSEBUTTONDOWN:
                #Player 1's Turn
                if turn == 0: 
                    col = drop_down(event.pos[0])
                    if is_valid_location(col) and open_row(board, col):
                        board[row, col] = 1;
                    draw_board(board)
                    pygame.display.update()
                    print(board)
                    win(board, row, col,turn)
                    turn = 1; 

                #Player 2's Turn
                elif turn == 1:
                        col = drop_down(event.pos[0])
                        if is_valid_location(col) and open_row(board, col):
                            board[row, col] = 2;
                        print(board)
                        draw_board(board)
                        pygame.display.update()
                        win(board, row, col,turn)
                        turn = 0; 

    pygame.quit()
```

```python
#initalize pygame before any use of pygame
pygame.init()
screen = pygame.display.set_mode(size)
main_menu()
```
